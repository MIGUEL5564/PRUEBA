# Ecuaciones Diferenciales

Las ecuaciones diferenciales son ecuaciones que involucran **derivadas** de una o más funciones desconocidas. Son herramientas fundamentales en muchas áreas de la ciencia, como la física, la biología, la ingeniería y la economía, ya que modelan fenómenos dinámicos.

---

## Definición de Ecuación Diferencial

Una **ecuación diferencial** es una ecuación que relaciona una función desconocida con sus derivadas. Estas derivadas representan las tasas de cambio de la función con respecto a sus variables independientes.

Por ejemplo, una ecuación diferencial de primer orden es:

$$
\frac{dy}{dx} = x + y
$$

---

## Tipos de Ecuaciones Diferenciales

Existen diferentes tipos de ecuaciones diferenciales, principalmente:

### 1. Ecuaciones Diferenciales Ordinarias (EDO)

Una **ecuación diferencial ordinaria** (EDO) es aquella que involucra una sola variable independiente. Es decir, la derivada de una función es respecto de una única variable. 

Ejemplo:

$$
\frac{dy}{dx} = 3x^2 + 4y
$$

### 2. Ecuaciones Diferenciales Parciales (EDP)

Una **ecuación diferencial parcial** (EDP) involucra derivadas parciales de una función con respecto a varias variables independientes.

Ejemplo:

$$
\frac{\partial u(x,y,z)}{\partial t} = k \frac{\partial^2 u}{\partial x^2}
$$

---

## Orden de una Ecuación Diferencial

El **orden** de una ecuación diferencial es el mayor grado de la derivada presente en la ecuación. Es decir, si la ecuación contiene una derivada de orden \( n \), se dice que es una ecuación diferencial de orden \( n \).

### Ejemplo:

Para la ecuación:

$$
\frac{d^2 y}{dx^2} + \frac{dy}{dx} = 0
$$

El orden es 2, ya que la derivada de mayor orden es $$ \frac{d^2 y}{dx^2} .$$

---

## Ejemplos de Ecuaciones Diferenciales

A continuación, presentamos 5 ejemplos de ecuaciones diferenciales para ilustrar las distintas formas y órdenes:

1. **EDO de primer orden:**

   $$
   \frac{dy}{dx} = 2x + 3y
   $$

2. **EDO de segundo orden:**

   $$
   \frac{d^2 y}{dx^2} - 4y = 0
   $$

3. **EDP de primer orden:**

   $$
   \frac{\partial u}{\partial t} = k \frac{\partial^2 u}{\partial x^2}
   $$

4. **EDO de primer orden:**

   $$
   \frac{dy}{dx} = \frac{3x + 1}{y}
   $$

5. **EDP de segundo orden:**

   $$
   \frac{\partial^2 u}{\partial t^2} = c^2 \left( \frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} \right)
   $$

Para los siguientes ejercicios identificar y clasificar ecuaciones diferenciales:

$$1. \frac{d^2 y}{dx^2} + y = 0 $$

$$2. \frac{dy}{dx} = 2x + y $$

$$3. \frac{\partial u}{\partial x} + \frac{\partial u}{\partial y} = 0 $$

$$4. \frac{d^3 y}{dx^3} - 6y = 0 $$

$$5. \frac{\partial^2 u}{\partial t^2} - c^2 \frac{\partial^2 u}{\partial x^2} = 0 $$

$$6. \frac{dy}{dx} = \frac{y}{x^2 + 1} $$

$$7. \frac{\partial u}{\partial t} = \frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} $$

## Conclusión

Las ecuaciones diferenciales son esenciales para modelar sistemas dinámicos. Aprender a identificarlas y clasificarlas es el primer paso para entender cómo resolverlas y aplicarlas en diversas disciplinas.
